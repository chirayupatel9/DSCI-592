```python
import torch

# Print allocated memory (used by tensors)
print(f"Allocated Memory: {torch.cuda.memory_allocated() / (1024*1024):.2f} MB")

# Print reserved memory (reserved by CUDA for future needs)
print(f"Reserved Memory: {torch.cuda.memory_reserved() / (1024*1024):.2f} MB")

# Print total memory (device total)
total_memory = torch.cuda.get_device_properties(0).total_memory
print(f"Total GPU Memory: {total_memory / (1024*1024):.2f} MB")

# Print memory summary
print(torch.cuda.memory_summary())
```
